```python
# Address missing data
```

I've chosen not to act on the missing data, since the fact that this column has no data is significant. Replacing with another number would be misleading (ex: replacing with 0 would lead to conclusions that clients placed a new order on the same day) or impact future calculations; I also don't want to filter these values out, as they are not really missing. With that in mind, I've creates a new variable to act as a flag (ex: 'first_order_flag' with values True/False).

```python
## - Adding a flag column 'first_order_flag'

df_ords['first_order_flag'] = df_ords['days_since_prior_order'].apply(lambda x: 'True' if pd.isna(x) else 'False')
```
