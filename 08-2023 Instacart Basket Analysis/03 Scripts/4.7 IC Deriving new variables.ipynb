### 1a. If-Statements

```python
# Define function to categorise items into price ranges

def price_label (row):
    
    if row['prices'] <= 5:
        return 'Low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid range product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return np.nan #else will catch missing values
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts (dropna = False)
```

```python
# Check max price to confirm
df['prices'].max()
```

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

### 1c. If-Statements with For-Loops

```python
# Checking number of orders per day of the week

ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Applying an if-statement with for-loops

result = []

for value in ords_prods_merged['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result. append('Regularly busy')
```

## 2a. Categorise into 'busiest days'

```python
# Busiest days = 0 and 1; Slowest days = 3 and 4; Regular days = else
```

```python
# Applying an if-statement with for-loops

result_2 = []

for value_2 in ords_prods_merged['orders_day_of_week']:
    if value_2 in [0, 1]:
        result_2.append('Busiest days')
    elif value_2 in [3, 4]:
        result_2.append('Slowest days')
    else:
        result_2.append('Regularly busy days')


        
# Combine with ords_prods_merged. Set a new column within ords_prods_mergde and set it to result_2

ords_prods_merged['busiest_days_grouped'] = result_2


# Check frequencies

ords_prods_merged['busiest_days_grouped'].value_counts(dropna = False)
```

## 2b. Categorise into 'busiest periods of the day'

```python
# Identifying the busiest hours of the day
ords_prods_merged['time_of_order'].value_counts(dropna = False)
```

```python
# 'Most orders'= 10,11,14,15,13,12 (>= 2500000 orders); 'Fewest orders' = 23,6,0,1,5,2,4,3 (<= 500000 orders); 'Average orders' = else
#'Most orders' = 10 to 15, 'Fewest orders' = 0 to 6 and 23
```

```python
# Applying an if-statement with for-loops

result = []

for value in ords_prods_merged['time_of_order']:
    if value >= 10 and value <= 15:
        result.append('Most orders')
    elif value >= 0 and value <= 6 or value == 23:
        result.append('Fewest orders')
    else:
        result.append('Average orders')
```

```python
# Combine with ords_prods_merged. Set a new column within ords_prods_mergde and set it to result_2

ords_prods_merged['busiest_period_of_day'] = result
```

```python
# Check frequencies

ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```

```python
# Checking for accuracy
ords_prods_merged['busiest_period_of_day'].shape  # Total days matches rows
```
