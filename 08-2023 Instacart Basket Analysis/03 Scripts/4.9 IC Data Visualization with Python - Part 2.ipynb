### This script contains the following points:
   1. Bar chart 'orders_day_of_week'
   2. Histograms 'prices'
   3. Scatterplot 'prices'
   4. Line chart 'orders_day_of_week' vs 'prices' (sample data)

```python
# Import libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Import dataframe
path = r'C:\Users\bruna\Career Foundry\08-2023 Instacart Basket Analysis'

ords_prods_all = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_all.pkl'))
```

### 1. Bar chart 'orders_day_of_week'

```python
bar = ords_prods_all['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'bar_orders_dow.png'))
```

```python
bar = ords_prods_all['orders_day_of_week'].value_counts().sort_index().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'bar_orders_dow_colour.png'))
```

 ### 2. Histograms 'prices'

```python
hist = ords_prods_all['prices'].plot.hist(bins = 70)
```

```python
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'hist_prices.png'))
```

### 3. Scatterplot 'prices'

```python
sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_all)
```

### 4. Line chart 'orders_day_of_week' vs 'prices' (sample data)

```python
np.random.seed(4)
dev = np.random.rand(len(ords_prods_all)) <= 0.7
```

```python
# store 70% of the sample in the dataframe big and 30% in dataframe small
big = ords_prods_all[dev]
small = ords_prods_all[~dev]
```

```python
# Double check length of original dataframe vs both samples together
len(ords_prods_all)
```

```python
len(big) + len(small)
```

```python
# Reducing the sample to only the columns needed for the chart
df_2 = small[['orders_day_of_week','prices']]
```

```python
# Creating the chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
line.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'line_prices_dow.png'))
```

### 5. Histogram 'time_of_order'

```python
hist_time_order = ords_prods_all['time_of_order'].plot.hist(bins = 12)
plt.xticks(range(0, 25, 2)) 
plt.show()
```

```python
hist_time_order.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'hist_time_order.png'))
```

```python
# The  big majority of orders happen between 10am and 5pm. After 5pm the number of orders placed starts decreaing, but there is still a significant number of orders between 5pm and 9pm. From then onwards, this number is low, which almost no sales recorded between midnight and 6 am.
```

### 6. Line chart 'prices' vs 'time_of_order'

```python
# creating a sample to only the columns needed for the chart (using small sample)
df_3 = small[['time_of_order','prices']]
```

```python
# creating the chart
line_prices_time = sns.lineplot(data = df_3, x='time_of_order',y = 'prices')
plt.xticks(range(0, 25, 2))
plt.show()
```

```python
# overall there isn't a big difference in the average price paid per order throughout the day
```

```python
line_prices_time.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'line_prices_time.png'))
```

### 7. Line chart 'age' vs 'nr of dependants'

```python
# creating a sample to only the columns needed for the chart (using small sample)
df_4 = small[['age','nr_dependants']]
```

```python
# creating the chart
line_age_dependts = sns.lineplot(data = df_4, x='age',y = 'nr_dependants')
```

```python
# there isn't a direct correlation between age and number of dependendants.
```

```python
line_age_dependts.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'line_age_dependts.png'))
```

### 8. Scatterplot 'age' vs 'income'

```python
sct_age_income = sns.scatterplot(x = 'age', y = 'income',data = ords_prods_all)
```

```python
# the majority of users habe an income under 200000 USD, with the biggest concentration of salaries falling between this number accross all ages. Users under over 40 years old The top salary for under 40s is at 400000USD. Users over 40, on the other hand have a bigger range of salaries, with a significant numebr of users with salaries between 200000 and 300000 USD, and the top salaries sitting at 600000 USD.
```

```python
sct_age_income.figure.savefig(os.path.join(path, '04 Analysis','Visualisations', 'sct_age_income.png'))
```
